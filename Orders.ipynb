```python
import pandas as pd

# Assuming 'orderDate' and 'shippedDate' are in datetime format
df['orderDate'] = pd.to_datetime(df['orderDate'])
df['requiredDate'] = pd.to_datetime(df['requiredDate'])
df['shippedDate'] = pd.to_datetime(df['shippedDate'])

# Fill missing values in 'shippedDate' by adding 7 days to 'orderDate' only for NaN values
df.loc[df['shippedDate'].isna(), 'shippedDate'] = df.loc[df['shippedDate'].isna(), 'orderDate'] + pd.Timedelta(days=7)

df
```
